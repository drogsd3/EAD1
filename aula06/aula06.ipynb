# Comandos Condicionais

## If, Else e Elif

### Exemplos

```python
idade = int(input("Qual sua idade: "))
if idade >= 18:
    print("Pode dirigir!")
else:
    print("Chama Uber!")
```

```python
idade = int(input("Digite sua idade: "))
carro = input("Tem carro? <S/N>")
if idade >= 18:
    if carro == "S":
        print("Pode dirigir!")
    else:
        print("Chama o Uber")
else:
    print("Chama o Uber")
```

```python
idade = int(input("Digite sua idade: "))
carro = input("Tem carro? <S/N>")
if (idade >= 18) and (carro == "S"):
        print("Pode dirigir!")
else:
    print("Chama o Uber")
```

```python
idade = int(input("Digite sua idade: "))
if (idade >= 18):
    carro = input("Tem carro? <S/N>")
    if carro == "S":
        print("Pode dirigir!")
    else:
        print("Chama o Uber")
else:
    print("Chama o Uber")
```

```python
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
media = (nota1+nota2)/2
print("A media foi: ", media)

if media >= 5:   
    print("Aprovado")
else:
    print("Reprovado")
```

```python
nota1 = float(input("Digite a primeira nota: "))
if nota1 > 7:
    nota1 += 0.5
nota2 = float(input("Digite a segunda nota: "))
if nota2 > 7:
    nota2 += 0.5
media = (nota1+nota2)/2
print("A media foi: ", media)

if media >= 5:   
    print("Aprovado")
else:
    print("Reprovado")
```
