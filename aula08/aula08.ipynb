```python
import random
```

```python
def validar_numero():
    while True:
        escolha = int(input("Digite um numero inteiro entre 1 a 50:"))
        if escolha < 1 or escolha > 50:
            print("Numero Invalido\nEscolha um numero entre 1 e 50:")
        else:
            return escolha
```

```python
sorteio = random.randint(1,50)

while True:
    escolha = validar_numero()
    if escolha == sorteio:
        print("Parabéns você acertou!!!\nO Numero era ", sorteio)
        break
    else:
        print("Você errou! :(")
        if(sorteio < escolha):
            print("O numero sorteado é menor que do que você escolheu\nTente novamente")
        else:
            print("O numero sorteado é maior do que você escolheu\nTente novamente")
        
```
