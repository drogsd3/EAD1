```python
import random
```

```python
def validar_numero(tentativas):
    while True:
        print(f"Tentativa {tentativas} de 5")
        escolha = int(input("Digite um numero inteiro entre 1 a 50:"))
        if escolha < 1 or escolha > 50:
            print("Numero Invalido\n")
        else:
            return escolha
```

```python
sorteio = random.randint(1,50)
tentativas = 0

while tentativas < 5:    
    escolha = validar_numero(tentativas)
    if escolha == sorteio:
        print("Parabéns você acertou!!!\nO Numero era ", sorteio)
        break
    else:
        print("Você errou! :(")
        if(sorteio < escolha):
            print("O numero sorteado é menor que do que você escolheu\nTente novamente")
        else:
            print("O numero sorteado é maior do que você escolheu\nTente novamente")
        tentativas += 1
        
print("Numero de tentativas excedida")
print("O numero sorteado era", sorteio)
        
```
